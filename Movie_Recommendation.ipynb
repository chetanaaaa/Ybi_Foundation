```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Movies%20Recommendation.csv')
```

```python
features=movies[['Movie_Genre','Movie_Keywords', 'Movie_Language','Movie_Cast','Movie_Director']].fillna(' ')
```

```python
features.shape
```

```python
features
```

```python
x=features['Movie_Genre']+' '+features['Movie_Keywords']+' '+features['Movie_Cast']+' '+features['Movie_Language']+' '+features['Movie_Director']
x.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
model=TfidfVectorizer()
```

```python
x=model.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(x)
similarity_score
```

```python
similarity_score.shape
```

```python
Fav_muv=input("Enter your favourite movie name: ")
```

```python
all_movies=movies['Movie_Title'].tolist()
```

```python
import difflib
```

```python
movie_Recommendation=difflib.get_close_matches(Fav_muv,all_movies)
movie_Recommendation
```

```python
close_match=movie_Recommendation[0]
close_match
```

```python
from time import clock_settime_ns
index=movies[movies.Movie_Title==close_match]['Movie_ID'].values[0]
print(index)
```

```python
Recommendation_Score=list(enumerate(similarity_score[index]))
print(Recommendation_Score)
```

```python
len(Recommendation_Score)
```

```python
Sorted_Similar_Movies=sorted(Recommendation_Score,key=lambda x:x[1],reverse=True)
print(Sorted_Similar_Movies)
```

```python
print('Top 30 Movies Suggested for You: \n ')
i=1
for movie in Sorted_Similar_Movies:
  index=movie[0]
  title=movies[movies.index==index]['Movie_Title'].values[0]
  if(i<31):
    print(i,'.',title)
    i+=1
```

```python
fav=input("Enter your favourite film: ")
lst=movies["Movie_Title"].tolist()
close=difflib.get_close_matches(fav,lst)
close1=close[0]
index=movies[movies.Movie_Title==close1]['Movie_ID'].values[0]
Recommendation_Score=list(enumerate(similarity_score[index]))
sort=sorted(Recommendation_Score,key=lambda x:x[1],reverse=True)
print("Top 10 movies suggested:\n")
i=1
for movie in sort:
  index=movie[0]
  title=movies[movies.index==index]['Movie_Title'].values[0]
  if(i<11):
    print(i,'.',title)
    i+=1
```
